```python
country_code=pd.read_excel("Country-Code.xlsx") 
```

```python
zomato=zomato.merge(country_code,on="Country Code",how="left") #merging the data of country code with main data
```

```python
plt.pie(zomato['Country'].value_counts().head(3),labels=zomato["Country"].value_counts().head(3).index,autopct="%1.2f%%")
```

```python
ratings =zomato.groupby(["Aggregate rating",'Rating color','Rating text']).size().reset_index() # creating ratings dataframe
```

```python
ratings.rename(columns={0:'Counts'},inplace=True)
```

```python
ratings
```

```python
# plt.bar(x="Aggregate rating",y="Rating Count",data=ratings)
plt.figure(figsize=(12,6))
# sns.barplot(x="Aggregate rating",y="Counts",data=ratings)
sns.barplot(x="Aggregate rating",y="Counts",data=ratings)
```

```python
#0 rating are highest and rating from 2.5 to 4.5 are somewhat high it means AVG rating are high
```
